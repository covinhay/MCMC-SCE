```python
import scipy.stats as stats

# sample data
observations = [26, 20, 30, 22, 33, 19]

print("Total Rolls:", sum(observations))
```

```python
# expected observations assuming a fair die
expected = [25, 25, 25, 25, 25, 25] 

# calculate p_value from Pearson's chi-squared test (goodness-of-fit test)
chi, p_value = stats.chisquare(f_obs = observations, f_exp = expected)

print(f"p-value: {p_value:.3f}")
```
