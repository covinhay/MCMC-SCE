```python
# Created by HEC on 01/29/2026
# first example adapted from GeeksforGeeks: https://www.geeksforgeeks.org/machine-learning/markov-chain/
import numpy as np
import scipy.linalg
```

```python
# trasnition matrix gives the probabilities for transitio from one state to another 

states = ["A", "B"]
trans_matrix = np.array([[0.6, 0.4],
                        [0.7, 0.3]])

# row 0 ([0.6, 0.4]) means startin at "A" with 60% chance to stay in "A", 40% to move to "B"
# row 1 ([0.7, 0.3]) means starting at "E" with 70% chance to move to "A", 30% to stay in "B"
```

```python
# set # of steps in walk
n_steps = 15

# start at "A"
current_state = 0 

# format start of chain
print(states[current_state], end=" -> ")

# np.random.choice picks next state (0/"A" or 1/"B") in chain based on trans_matrix of probabilities
for i in range(n_steps - 1):
    current_state = np.random.choice(
                    [0, 1], p = trans_matrix[current_state])
    print(states[current_state], end=" -> ")

# format end of chain
print("stop")
```

```python
# stationary distribtion gives the long-run probability of being in each state

# compute left eigenvectors of trans_matrix
# stationary distribution corresponds to the eigenvector associated with eigenvalue 1

eigvals, left_eigvecs = scipy.linalg.eig(
                        trans_matrix, left=True, right=False)

# disregard any imaginary component
stationary = left_eigvecs[:, 0].real

# normalizes vector so probabilities/entries sum to 1
stationary /= stationary.sum()

print("Stationary distribution:", stationary)
```
