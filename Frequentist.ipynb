```python
import random 

sides = 6
rolls = []

for i in range(200):
    roll = random.randint(1, sides)
    rolls.append(roll)

print(rolls)
```

```python
print("\nCounts:")
for i in range(1, sides+1):
    print(f"{i}: {rolls.count(i)}")
```

```python
print("\nPercentages:")
for i in range(1, sides+1):
    print(f"{i}: {float(rolls.count(i)/len(rolls)*100): .2f}%")
```
