```python
import scipy.stats as stats
import matplotlib.pyplot as plt

n = 10 # number of trials

p = 0.2 # probability of success

x = list(range(0, n+1)) # all possible values X can take

pmf = stats.binom.pmf(x, n, p) # computes P(X=x) for each x

plt.figure(figsize = (10, 5))
plt.bar(x, pmf, color = "blue", width = 0.6)
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
```

```python
import numpy as np

mu = 0

sigma2 = 3

x = np.linspace(-6, 6, 1000)

pdf = stats.norm.pdf(x, mu, sigma2)

plt.figure(figsize = (10, 5))
plt.plot(x, pdf, color = "b")
plt.title(f"N({mu}, {sigma2})")
plt.xlabel("x")
plt.ylabel("Probability Density")
```

```python
# plot uniform distribution

a = 0
b = 1

x = np.linspace(0, 1, 1000)

pdf = stats.uniform.pdf(x, a, b-a)

plt.figure(figsize = (10, 5))
plt.plot(x, pdf, color = "b")
plt.title(f"U({a}, {b})")
plt.xlabel("x")
plt.ylabel("Probability Density")
```

```python
# plt beta distribution

a, b = 1, 1 # shape parameters 

x = np.linspace(0, 1, 1000)

pdf = stats.beta.pdf(x, a, b)

plt.figure(figsize = (10, 5))
plt.plot(x, pdf, color = "b")
plt.title(f"Beta({a}, {b})")
plt.xlabel("x")
plt.ylabel("Probability Density")
```
