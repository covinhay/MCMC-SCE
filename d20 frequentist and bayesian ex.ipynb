```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
# calculate p-value from binomial dsitribution

n       = 200 # number of rolls

p       = 1/20 # prob of rolling critical hit with fair die

k       = 19 # number of critical hits

p_value = 1 - stats.binom.cdf(k -1, n, p) # computes 1 - P(X < k) = P(X >= k)

print(f"p-value: {p_value:.4f}")
```

```python
# plot binomial distribution w/ n=200 and p=1/20

x = list(range(0, n+1)) # all possible values X can take

pmf = stats.binom.pmf(x, n, p) # computes P(X = x) for each x

plt.figure(figsize = (15, 5))
plt.bar(x, pmf, color = "blue", width = 0.6)
plt.xlabel("Number of 20s")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
```

```python
plt.figure(figsize = (15, 5))
plt.bar(x, pmf, color = "blue", width = 0.6)
plt.xlabel("Number of 20s")
plt.ylabel("Probability")
plt.yscale("log")
plt.title(f"Binomial Distribution (n={n}, p={p}) - Log Scale")
```

```python
# plot binomial distirbution w/ p-value visualized

colors  = ["r" if i >= k else "b" for i in x] # red bars for P(X >= k) and blue bars for P(X < k)

# custom legend content
blue    = patches.Patch(color = "blue", label = f"P(X < {k})")
red     = patches.Patch(color = "red", label = f"P(X >= {k})")

# plot
plt.figure(figsize = (15, 5))
plt.bar(x, pmf, color = colors, width = 0.6) 
plt.legend(handles = [blue, red]) # print custom legend on plot
plt.xlabel("Number of 20s")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
```

```python
# plot with p-value visualized on log scaled y-axis

plt.figure(figsize = (15, 5))
plt.bar(x, pmf, color = colors, width = 0.6) 
plt.legend(handles = [blue, red]) # print custom legend on plot
plt.xlabel("Number of 20s")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
```

```python
# bayesian model

a, b = 1, 1 # prior

k, j = 16, 184 # observed successes and failures 

prob = 1- stats.beta.cdf(0.055, a + k, b + j) #P(p >= 0.055)

print(prob)
```

```python
# plot beta model
x = np.linspace(0, 1, 1000)

pdf = stats.beta.pdf(x, a + k, b + j)

plt.figure(figsize = (10, 5))
plt.plot(x, pdf, color = "b")
plt.title(f"Beta({a+k}, {b+j})")
plt.xlabel("x")
plt.ylabel("Probability Density")
```
